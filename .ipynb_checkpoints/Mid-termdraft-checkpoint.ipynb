# Merge two datasets 
We merge the two datasets 2023_Eviction_Notices.csv and evictions_geocoded.csv 

```python
# Extracting the name from the input_string in df1
df_geocoded['Name'] = df_geocoded['input_string'].apply(lambda x: x.split(',')[0].strip())

# Merging on the 'Name' column
merged_df = pd.merge(df_geocoded, df_evictions, left_on='Name', right_on='Address', how='inner')

# Displaying the merged dataframe
merged_df.head()
```

```python
import geopandas as gpd
import pandas as pd
# Load the datasets
df_evictions = gpd.read_csv('data/2023_Eviction_Notices.csv')
#df_geocoded = gpd.read_csv('evictions_geocoded.csv')
df_geocoded = pd.read_csv('data/evictions_geocoded.csv')

# Convert them to GeoDataFrames if they are not already
gdf_evictions = gpd.GeoDataFrame(df_evictions, geometry=gpd.points_from_xy(df_evictions.City, df_evictions.Zip))
gdf_geocoded = gpd.GeoDataFrame(df_geocoded, geometry=gpd.points_from_xy(df_geocoded.latitude, df_geocoded.longitude))

# Perform the spatial join
spatial_join = gpd.sjoin(gdf_evictions, gdf_geocoded, how='inner', op='intersects')

# Display the spatially joined DataFrame
print(spatial_join)
```

```python
import geopandas as gpd
import pandas as pd

# Load the datasets
df_evictions = gpd.read_file('data/2023_Eviction_Notices.csv')
df_geocoded = pd.read_csv('data/evictions_geocoded.csv')

# Convert them to GeoDataFrames if they are not already
gdf_evictions = gpd.GeoDataFrame(df_evictions, geometry=gpd.points_from_xy(df_evictions.City, df_evictions.Zip))
gdf_geocoded = gpd.GeoDataFrame(df_geocoded, geometry=gpd.points_from_xy(df_geocoded.latitude, df_geocoded.longitude))

# Perform the spatial join
spatial_join = gpd.sjoin(gdf_evictions, gdf_geocoded, how='inner', op='intersects')

# Display the spatially joined DataFrame
print(spatial_join)
```
