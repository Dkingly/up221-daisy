## Latitude and longitude  identifying WA 

```python
# latitude and longitude  identifying WA 
#latitude = 47.751076
#latitude = -120.740135
import folium

# Latitude and Longitude for Washington (WA)
wa_lat, wa_lon = 47.7511, -120.7401

#map centered around WA
map_wa = folium.Map(location=[wa_lat, wa_lon], zoom_start=8)

# We add marker for WA
folium.Marker(location=[wa_lat, wa_lon], popup='Washington').add_to(map_wa)


# Save the map as an HTML file
map_wa.save('washington_map.html')

# Display the map
map_wa
```

## Latitude and longitude identifying WA and the Counties 

```python
import folium
import pandas as pd

# County data with major cities
county_data = {
    "County": ["King", "Clark", "Kitsap", "Snohomish", "Thurston", "Cowlitz", "Jefferson", "Yakima", "Island",
               "Stevens", "Skagit", "Clallam", "Chelan", "Walla Walla", "Kittitas", "Spokane", "Klickitat", "Whitman", "Grant"],
    "City": ["Seattle", "Vancouver", "Bremerton", "Everett", "Olympia", "Longview", "Port Townsend", "Yakima", "Coupeville",
             "Colville", "Mount Vernon", "Port Angeles", "Chelan", "Walla Walla", "Ellensburg", "Spokane", "Goldendale", "Colfax", "Ephrata"],
    "Lat": [47.6062, 45.6387, 47.5673, 47.9789, 47.0379, 46.1417, 48.1170, 46.6021, 48.2192, 48.5459, 48.4249, 48.1181, 47.8379, 46.0679, 47.0006, 47.6588, 45.8209, 46.8721, 47.2885],
    "Lon": [-122.3321, -122.6615, -122.6256, -122.2021, -122.9007, -122.9365, -122.7605, -120.5059, -122.6856, -117.8864, -122.3157, -123.4315, -120.0250, -118.3430, -120.5465, -117.4260, -120.7401, -117.1582, -119.2872]
}

df_county = pd.DataFrame(county_data)

# Create a map centered around WA
map_wa_city = folium.Map(location=[47.6062, -120.7401], zoom_start=7)

# Add markers for major cities in each county
for index, row in df_county.iterrows():
    folium.Marker(
        location=[row["Lat"], row["Lon"]],
        popup=row["City"],
        icon=folium.Icon(color='blue')
    ).add_to(map_wa_city)

# Save the map as an HTML file
map_wa_city.save('washington_cities_map.html')

# Display the map
map_wa_city
```
